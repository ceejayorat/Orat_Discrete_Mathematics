# 1 . Sets and Subsets

1. **Proving A\B = A $\cap$ $B^{c}$**

   To prove this identity, we'll show that the sets on both sides of the equation are equal.
 
   \begin{align*}
   A \setminus B &= A \cap B^{c} \\
   &= \{x \mid x \in A \text{ and } x \notin B\} \\
   &= \{x \mid x \in A \text{ and } x \in B^{c}\} \\
   &= A \cap B^{c}
   \end{align*}

2. **Proving B = (B $\cap$ A) $\cup$ (B $\cap$ $A^{c}$)**

   We will prove the set equality step by step.

   \begin{align*}
   B &= (B \cap A) \cup (B \cap A^{c}) \\
   &= \{x \mid x \in B \text{ and } (x \in A \text{ or } x \in A^{c})\} \\
   &= \{x \mid (x \in B \text{ and } x \in A) \text{ or } (x \in B \text{ and } x \in A^{c})\} \\
   &= \{x \mid x \in B \cap A \text{ or } x \in B \cap A^{c}\} \\
   &= (B \cap A) \cup (B \cap A^{c})
   \end{align*}

3. **Proving B\A = B $\cap$ $A^{c}$**

   We will demonstrate that the two sets are equal.

   \begin{align*}
   B \setminus A &= B \cap A^{c} \\
   &= \{x \mid x \in B \text{ and } x \notin A\} \\
   &= \{x \mid x \in B \text{ and } x \in A^{c}\} \\
   &= B \cap A^{c}
   \end{align*}

4. **Proving A $\cup$ B = A $\cup$ (B $\cap$ $A^{c}$)**

   We'll establish the equality of the sets using logical reasoning.

   \begin{align*}
   A \cup B &= A \cup (B \cap A^{c}) \\
   &= \{x \mid x \in A \text{ or } (x \in B \text{ and } x \in A^{c})\} \\
   &= \{x \mid (x \in A \text{ or } x \in B) \text{ and } (x \in A \text{ or } x \in A^{c})\} \\
   &= \{x \mid x \in A \text{ or } (x \in B \text{ and } x \notin A)\} \\
   &= \{x \mid x \in A \text{ or } x \in B \setminus A\} \\
   &= A \cup B \setminus A \\
   &= A \cup (B \cap A^{c})
   \end{align*}

5. **Proving $(\textbf{U}_{i=1}^{\infty} A_{i}^{c})^{c}$ = $\cap_{i=1}^{\infty}A_{i}$**

   We'll use set complement properties and De Morgan's law to establish the equivalence.

   \begin{align*}
   \left(\bigcup_{i=1}^{\infty} A_{i}^{c}\right)^{c} &= \bigcap_{i=1}^{\infty} A_{i} \\
   &= \{x \mid x \notin \bigcup_{i=1}^{\infty} A_{i}^{c}\} \\
   &= \{x \mid x \notin A_{i}^{c} \text{ for all } i\} \\
   &= \{x \mid x \in A_{i} \text{ for all } i\} \\
   &= \bigcap_{i=1}^{\infty} A_{i}
   \end{align*}

6. **Proving |A $\cup$ B| = |A| + |B| - |A $\cap$ B|**

   This identity is a result of the Inclusion-Exclusion Principle.

   \begin{align*}
   |A \cup B| &= |A| + |B| - |A \cap B|
   \end{align*}

1. **C symmetric difference D:**
   The symmetric difference between two sets A and B, denoted as AΔB, is the set of elements that are in either A or B, but not in both.
   
   \(CΔD=(C∪D)−(C∩D)\)
   
   \(C∪D=\{5,6,7,8,9\}\)
   
   \(C∩D=\{5,7\}\)
   
   \(CΔD=\{6,8,9\}\)

2. **A\B (Set difference A minus B):**
   This is the set of elements that are in A but not in B.
   
   \(A∖B=\{1,2,3\}\)

3. **B\A (Set difference B minus A):**
   This is the set of elements that are in B but not in A.
   
   \(B∖A=\{6,7\}\)

4. **E direct sum F:**
   The direct sum of two sets E and F is the set of all possible ordered pairs where the first element is from E and the second element is from F.
   
   \(E⊕F=\{(2,1),(2,5),(2,9),(4,1),(4,5),(4,9),(6,1),(6,5),(6,9),(8,1),(8,5),(8,9)\}\)

With these information do the following:<br>

**2. Find the number of people who read at $\textbf{"least one"}$ of the three magazines.**

   *To find the number of people who read at least one magazine, we can use the principle of inclusion-exclusion: Number of people who read at least one magazine = N + T + F - (NT + NF + TF) + NTF*

**3. Find the number of people who read exactly one magazine.**
<br>

   *To find the number of people who read exactly one magazine, we need to sum up the counts of people who read each magazine exactly, i.e., subtracting the overlaps and the count of those who read all three magazines: Number of people who read exactly one magazine = (N - NT - NF + NTF) + (T - NT - TF + NTF) + (F - NF - TF + NTF)*

```python
from matplotlib_venn import venn3
import matplotlib.pyplot as plt

# Given data
total_people = 120
newsweek_readers = 65
time_readers = 45
fortune_readers = 42
newsweek_time_readers = 20
newsweek_fortune_readers = 25
time_fortune_readers = 15
all_three_readers = 8

# Calculate the number of people who read at least one magazine
at_least_one = newsweek_readers + time_readers + fortune_readers - \
              newsweek_time_readers - newsweek_fortune_readers - \
              time_fortune_readers + all_three_readers

# Calculate the number of people who read exactly one magazine
exactly_one = (newsweek_readers - newsweek_time_readers - newsweek_fortune_readers + all_three_readers) + \
              (time_readers - newsweek_time_readers - time_fortune_readers + all_three_readers) + \
              (fortune_readers - newsweek_fortune_readers - time_fortune_readers + all_three_readers)

# Create Venn diagram
venn = venn3(subsets=(newsweek_readers - newsweek_time_readers - newsweek_fortune_readers + all_three_readers,
                      time_readers - newsweek_time_readers - time_fortune_readers + all_three_readers,
                      newsweek_time_readers - all_three_readers,
                      fortune_readers - newsweek_fortune_readers - time_fortune_readers + all_three_readers,
                      newsweek_fortune_readers - all_three_readers,
                      time_fortune_readers - all_three_readers,
                      all_three_readers),
             set_labels=('Newsweek', 'Time', 'Fortune'))
```

## B. In a survey of 25 new cars...

Given data:
- Total surveyed cars: 25
- Cars with A (Air-conditioning): 15
- Cars with R (Radio): 12
- Cars with W (Power windows): 11
- Cars with A and W: 5
- Cars with A and R: 9
- Cars with R and W: 4
- Cars with all three options: 3


1. Find the Number of Cars with W Only:
Number of cars with W only = W - (A and W) - (R and W) + (A, R, and W)

2. Find the Number of Cars with A Only:
Number of cars with A only = A - (A and W) - (A and R) + (A, R, and W)

3. Find the Number of Cars with R Only:
Number of cars with R only = R - (A and R) - (R and W) + (A, R, and W)

4. Find the Number of Cars with Both R and W but Not A:
Number of cars with both R and W but not A = (R and W) - (A, R, and W)

5. Find the Number of Cars with Both A and R but Not W:
Number of cars with both A and R but not W = (A and R) - (A, R, and W)

6. Find the Number of Cars with Only One of the Options:
Number of cars with only one option = (A - A, R, and W) + (R - A, R, and W) + (W - A, R, and W)

7. Find the Number of Cars with at Least One Option:
Number of cars with at least one option = (A or R or W) = Total surveyed cars - None of the Options

8. Find the Number of Cars with None of the Options:
Number of cars with none of the options = Total surveyed cars - (A or R or W)

```python
#1 
#Define the sets
A={15,5,9,3} #A
B={12,9,4,3} #R
C={11,5,4,3} #W

# Create the Venn diagram
venn = venn3(subsets=(len(A - B - C), len(B - A - C), len(A & B - C), len(C - A - B),
                      len(A & C - B), len(B & C - A), len(A & B & C)),
             set_labels=('A', 'R', 'W'))

# Add labels to the intersections
venn.get_label_by_id('100').set_text('4')
venn.get_label_by_id('010').set_text('2')
venn.get_label_by_id('110').set_text('6')
venn.get_label_by_id('001').set_text('5')
venn.get_label_by_id('101').set_text('2')
venn.get_label_by_id('011').set_text('1')
venn.get_label_by_id('111').set_text('3')

# Display the diagram
plt.title("Survey for 25 New Cars")
plt.show()
```
